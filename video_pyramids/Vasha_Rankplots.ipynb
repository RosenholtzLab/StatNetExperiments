```python
def plot_mean_rank(labels, values, dataset,transform):
    
    # Pastel colors
    #colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700', '#C2C2F0', '#FFB3E6', '#FF6666', '#C2C2F0']
    colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

    # Compute bar heights and bottoms based on 75 as the center
    bar_heights = [75 - val if val < 75 else val - 75 for val in values]
    bar_bottoms = [val if val < 75 else 75 for val in values]

    # Create the figure and axis with a reduced size for fitting multiple plots
    fig, ax = plt.subplots(figsize=(4, 4))

    # Plot bars with reduced width
    bar_width = 0.8
    for label, height, bottom, color in zip(labels, bar_heights, bar_bottoms, colors):
        ax.bar(label, height, width=bar_width, bottom=bottom, color=color, edgecolor='black', linewidth=0.5)

    # Setting the y-axis
    ax.set_ylim(150, 0)
    ax.axhline(75, color='black', linestyle='--', linewidth=bar_width)
    #ax.set_ylabel('Group Rank Average', fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=12)  # Updated labelsize here
    # Rotate the x-axis labels for better readability
    ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=13)

    # Remove y-axis grid lines and only keep the horizontal line at 75
    ax.yaxis.grid(False)

    # Remove the spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    titledict = {'HB':'Depleted',
                 'OT':'Original'}

    plt.title(f'{dataset} {titledict[transform]}')
    
    # Display the plot
    plt.tight_layout()  # Adjusts subplot params for better layout
    plt.savefig(f'./paper_plots/mean_rank_bar_{dataset}_{transform}.png', dpi=500)
    #plt.show()
    
```

```python
# Now, you can use these labels and values for plotting.
plot_mean_rank(labels, values, dataset, transform)
```

```python
import pickle
#dataset = 'DTD' #'KTH''DTD'
transform = 'OT' #'OT''HB

with open('/home/gridsan/ckoevesdi/StatNetExperiments/video_pyramids/averaged_feature_rankings_DTD_OT_shapley.pkl', 'rb') as file:
    dtd = pickle.load(file)
with open('/home/gridsan/ckoevesdi/StatNetExperiments/video_pyramids/averaged_feature_rankings_KTH_OT_shapley.pkl', 'rb') as file:
    kth = pickle.load(file)

kth= list(kth.values()) 
dtd = list(dtd.values())  
    
with open('/home/gridsan/ckoevesdi/StatNetExperiments/video_pyramids/group_avg_ranks_KTH_OT.pkl', 'rb') as file:
    loaded_group_avg_rank_KTH = pickle.load(file)
    
with open('/home/gridsan/ckoevesdi/StatNetExperiments/video_pyramids/group_avg_ranks_DTD_OT.pkl', 'rb') as file:
    loaded_group_avg_rank_DTD = pickle.load(file)

avg_KTH = list(loaded_group_avg_rank_KTH.values()) 
avg_DTD = list(loaded_group_avg_rank_DTD.values())    


# Groups of indices
rawstats = [0, 1]
bandpass_variance = [i for i, label in enumerate(ostat_labels) if 'bandpass_variance' in label]
low = [3, 4, 5]
nonoriented = list(range(6, 20))
edgemean = [i for i, label in enumerate(ostat_labels) if 'edge_mean' in label]
edgevariance = [i for i, label in enumerate(ostat_labels) if 'edge_variance' in label]
edgecorrelation = [i for i, label in enumerate(ostat_labels) if 'edge_correlation' in label]
phase = [i for i, label in enumerate(ostat_labels) if 'phase' in label]
stop = [i for i, label in enumerate(ostat_labels) if 'stop' in label]

labels_dtd = ["Marg", "BP Var", "LP Var", "Non Ori", "Mag Mn", "Mag Var", "Mag Cor", "Phs Cor", "End-Stp"]
labels_kth = ["Marg", "BP Var", "LP Var", "Non Ori", "Mag Mn", "Mag Var", "Mag Cor", "Phs Cor", "End-Stp"]
mean_values_dtd = avg_DTD
mean_values_kth = avg_KTH
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
#colors = np.array([plt.get_cmap('jet', 10).color])
#cmap = cm.get_cmap('tab10')
#colors = np.array([cmap(np.linspace(0, 1, 9))])
##colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700', '#C2C2F0', '#FFB3E6', '#FF6666', '#C2C2F0']
fig, ax = plt.subplots(figsize=(4, 4))  # increased the figure size slightly

for i, label in enumerate(labels_kth):
    idx_list_kth = groups[label]
    idx_list_dtd = groups[label]

    avg_ranks_kth = [kth[idx] for idx in idx_list_kth]
    avg_ranks_dtd = [dtd[idx] for idx in idx_list_dtd]

    plt.scatter(avg_ranks_kth, avg_ranks_dtd, c=colors[i], s=5)
    
for i in range(len(labels_kth)):
    plt.scatter(mean_values_kth[i], mean_values_dtd[i], c=colors[i], label=labels_kth[i], s=100, marker='o')

plt.xlabel('Ranking KTH', fontsize=12)  # increased fontsize
plt.ylabel('Ranking DTD', fontsize=12)  # increased fontsize

# Adjust the tick label font size
ax.tick_params(axis='both', which='major', labelsize=10)

plt.xlim(-3, 153)
plt.ylim(-3, 153)

# Adjust legend font size
plt.legend()

plt.tight_layout()
plt.savefig(f'./paper_plots/scatter_rank_shapley.png', dpi=500)

    
#plt.scatter(values_kth,values_dtd,c=colors)
```
