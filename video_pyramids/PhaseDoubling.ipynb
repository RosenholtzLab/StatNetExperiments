## Phase Doubling
Why the heck is there a 2x multiplyer in the phase?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.clf()

x1_limit = 5

x1 = np.linspace(-x1_limit,x1_limit)
s1 = np.sin(x1)
c1 = np.cos(x1)
plt.plot(x1,s1,label='sine')
plt.plot(x1,c1,label='cos')
plt.axvline(x=0,c='k')
plt.axhline(y=0,c='k')
plt.legend()
plt.title('Scale 1')
plt.show()

x2 = np.linspace(-2*x1_limit,2*x1_limit)
s2 = np.sin(x2/2)
c2 = np.cos(x2/2)
plt.plot(x2,s2,label='sine')
plt.plot(x2,c2,label='cos')
plt.axvline(x=0,c='k')
plt.axhline(y=0,c='k')
plt.legend()
plt.title('Scale 2')
plt.show()


angles_s1 = np.angle(c1 + (s1*1j))
angles_s2 = np.angle(c2 + (s2*1j))

plt.plot(x1,angles_s1,label='theta_scale1')
plt.plot(x1,angles_s2,label='theta_scale2')
plt.legend()
plt.title('Angles')
plt.show()

# double_angles = ((2*(angles_s1 + np.pi)) % (2*np.pi) ) - np.pi
# double_angles = 2*angles

# plt.plot(x,double_angles,label='2*theta_scale1')
# plt.legend()
# plt.title('Double Angles')
# plt.show()


```

```python
def draw_sine(scale,theta,c='k'):
    x = np.linspace(-scale*np.pi,scale*np.pi,100)
    y = scale*np.cos(x/scale + theta)
    plt.plot(x,y,label=f'scale={scale},theta={theta:0.2f}',c=c)
    yd = scale*np.sin(x/scale + theta)
    #plt.plot(x,yd,'.',label=f'scale={scale},theta={theta:0.2f} i',c=c)
    plt.plot(x,yd,'--',c=c)
    plt.axvline(x=0, color='k', linewidth=1.0, linestyle='-') 
    plt.axhline(y=0, color='k', linewidth=1.0, linestyle='-') 


angle = np.pi
    
draw_sine(1.,angle,c='b')
draw_sine(2.,angle,c='r')
draw_sine(2.,2*angle,c='g')

plt.legend()
plt.title(f'theta = {angle:0.2f}')
plt.show()
```
